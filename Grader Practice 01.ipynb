### Task P1_02 : Flowchart #02 (3 Star)




```python
# 1 : รับ input n,k (int) -> โจทย์ไม่บอก เเต่รู้เองว่ารับเเบบ list comprehension
n,k = [int(e) for e in input().split()]

# 2 : check condition ใหญ่ก่อน
if n % 2 != 0 :
    sum_a = 0 ; sum_b = 0 ; sum_c = 0 ; m = 0
    
    # วน while loop
    while m < k :
        a,b,c = [int(e) for e in input().split()]
    
        if a == b :
            if (a == b == c) :
                if (a + b) > k : # รอบคอบขึ้นหน่อย ระวังผิดngo x-x
                    sum_a += 1
                    sum_b += 2
                    sum_c -= 3
                else :           # รอบคอบขึ้นหน่อย ระวังผิดngo x-x
                    sum_a -= 3
                    sum_b -= 2
                    sum_c += 1
            else :
                sum_a += 2
                sum_b -= 3
        
        m += 1 # +1 ทุก loop
    
    # ออกจาก loop -> บังคับ print sum_a , sum_b , sum_c
    print(sum_a , sum_b , sum_c)
    
elif n % 2 == 0 :
    s,t = [int(e) for e in input().split()]
    
    x = s ; y = t
    
    if s > t :
        x = s - t
    
    else :
        if s < t : #เขียนย่อเป็น elif ก็ได้
            y = 2 * (t - s)
     
    # บังคับเช็ค if       
    if x+y > k :
        x,y = y,x
        x = y + s**2
    
    # บังคับ print x,y    
    print(x,y)
```

### Task P1_07 : RLE to String (2 Star)


```python
# 1 : รับ input เพื่อเเยกเคส
algo = str(input('Enter str2RLE or RLE2str : '))

# 2 : case 1 : str -> RLE
if algo == 'str2RLE' :
    # ใช้ Logic จากข้อ 04_08 RLE :
    
    # 2.1 : รับ input rle + สร้าง output str ว่าง (เดี๋ยวค่อยๆเพิ่ม str)
    rle1 = str(input('Enter RLEs : '))
    rle1_output = ''
    
    # 2.2 : ตั้งตป.ที่จำเป็น
    count1 = 1 ; check1 = rle1[0] # เจอตัวเเรก นับเป็น 1 , rle ให้เป็น str ตัวเเรก
    
    # 2.3 : for loop -> check cd.
    for i in range(1 , len(rle1)) : # เริ่ม check index 1 ถึงสุดท้าย
        if check1 == rle1[i] :
            count1 += 1
    
        elif check1 != rle1[i] :
        # เพิ่มค่าเดิมก่อน
            rle1_output += ( str(check1) + ' ' + str(count1) + ' ' ) # Ex : A 5 Y 3 
        # เปลี่ยนค่า count , check
            count1 = 1 # เจอตัวใหม่ นับเป็น 1
            check1 = rle1[i] #  เปลี่ยนให้เป็นตัวที่ i 
        
    # 2.4 : เพิ่มค่าสุดท้าย เเล้ว print rle_output
    rle1_output += ( str(check1) + ' ' + str(count1) + ' ' ) 
    print(rle1_output)
    
# 3 : RLE -> str
elif algo == 'RLE2str' :
    
    # 3.1 : split str ทั้งหมดก่อน +  สร้าง str ว่าง
    rle2 = input('Enter RLEs : ').split() # list of str
    str2_output = '' 
        
    # 3.2 : for loop อีกรอบ เพื่อจะทำการเพิ่มstr
    for j in range(0 , len(rle2) - 1 , 2) :
        str2_output += ( rle2[j] * int(rle2[j+1]) ) 
        # rle2[j] = ALphabet , rle2[j+1] = Number
        
    # 3.3 : print output
    print(str2_output)
    
# 4 : case อื่น
else :
    print('Error')
```

### Task P1_08 : Rock Scissors Paper (2 Star)


```python
# 1 : รับ input จน.เกมที่จะเล่น m (int)
m = int(input('Enter Round : '))

# 2 : สร้าง ตป.ที่จำเป็นต้องใช้
round = 0  
end = 3*m # 3 x จน.รอบ (ตามเงื่อนไขโจทย์X
tie = True # กำหนดว่าเสมอมั้ย (กำหนด default = True)
count_1 = 0; count_2 = 0

# 3 : while loop -> ตราบใดที่ยังไม่เกิน 3m รอบ เล่นต่อ
while round < end :
    # 3.1 : รับ input -> split เป็น list of str
    algo = input('ENter R/S/P by P1 P2 : ').split() # list of str
    
    # 3.2 : check เงื่อนไขที่ทำให้เเต่ละคนชนะ -> 6 cases (ตนละ 2 , 3 tools)
    if algo[0] == 'P' and algo[1] == 'R' :
        count_1 += 1
        
    elif algo[0] == 'P' and algo[1] == 'S' :
        count_2 += 1
        
    elif algo[0] == 'S' and algo[1] == 'P' :
        count_1 += 1
        
    elif algo[0] == 'S' and algo[1] == 'R' :
        count_2 += 1
        
    elif algo[0] == 'R' and algo[1] == 'P' :
        count_2 += 1
        
    elif algo[0] == 'R' and algo[1] == 'S' :
        count_1 += 1
    
    # 3.3 : เพิ่ม จน. round เมื่อเล่นเสร็จ
    round += 1
    
    # 3.4 : check cd -> ถ้า จน.ชนะ (count) = จน.รอบ (m) -> ให้คนนั้นชนะเลย + ออก loop
    if (count_1 == m):
        tie = False
        print(count_1 , count_2)
        print('Player 1 wins')
        break # ออกได้เลย ถ้าชนะ
    
    elif (count_2 == m) :
        tie = False
        print(count_1 , count_2)
        print('Player 2 wins')
        break # ออกได้เลย ถ้าชนะ
    
# 4 : จบ loop เเล้ว -> check ว่าเพราะอะไร (จบก่อน or Tie)
if tie : # ถ้า Tie -> print Tie & คะเเนน
    print(count_1 , count_2)
    print('Tie')
```

### Task P1_10 : Golf (2 Star)




```python
# 1 : สร้าง list ว่าง -> เพื่อเพิ่มค่าเข้าไปเรื่อยๆ ใน list
import math
par = [];  stroke = [];  choose = [];  repair = []

# 2 : for loop 9 รอบ : เพื่อรับ input & เพิ่มสมาชิก
for i in range(9) :
    info = input('Enter Par / Stroke / Choose : ')
    # 2.1 : split  เเล้วเเปลงสมาชิกเป็น intT
    p = int(info.split()[0]); s = int(info.split()[1]) ; ch = int(info.split()[2])
    
    # 2.2 : เพิ่มค่าเข้าไปเรื่อยๆ ใน list ว่าง
    par += [p];  stroke += [s];  choose += [ch] # Note : การนำ list + list = การนำสมาชิกใน list มาต่อกัน
    
    # ตอนนี้มีlist par , stroke , change , repair เเล้ว => list (of int)
    
    # 2.3 : check cd -> ถ้าป้อน 1 ให้ปรับปรุง stroke , ถ้าป้อน 0 ไม่ต้องทำอะไร
    if choose[i] == 1 :
        # stroke ปรับปรุง => เอา min ของค่า stroke ตัวนั้นๆ กับ ค่า par ตัวนั้นๆ +2
        change = min( stroke[i] , par[i] + 2 )
        repair += [change] # เพิ่มสมาชิกใน list

# 3 : ตอนนี้มี list ของทั้งหมดเเล้ว -> หา sum เเล้วเเทนเข้าสูตรหาเเต้มต่อ (extra)
par = sum(par);     stroke = sum(stroke);   repair = sum(repair);    # หาค่า sum เเล้วเเทนลง ตป.เดิม
    
extra = math.floor( 0.8 * (1.5 * repair - par) ) # เกิดปัญหาว่า ถ้าใช้ int มันจะปัดขึ้น (in case จน.ลบ) => เเก้โดยลองใช้ floor fn (ไม่เเนะนำ) หรืออาจจะต้องกำหนด condition ในการใช้
total = stroke - extra # int

# 4 : print stroke , extra , total คนละบรรทัด
print(stroke);  print(extra);   print(total)
```
